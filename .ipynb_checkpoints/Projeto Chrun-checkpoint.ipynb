# Análise de dados de uma empresa de telecomunicação

### Informação da base de dados
Nós trabalhamos em uma empresa de telecomunicações e um dos problemas da empresa é a alta taxa de churn. Para resolver isso, solicitamos uma base de dados com informações de diversos clientes anteriores para analisar e encontrar padrões que possam ajudara empresa.

E logo após irei criar um dashboard para acompanhamento das principais informações.

### Informação da base de dados
- <b>customerID:</b> Código de identificação único do cliente
- <b>gender:</b> Gênero do cliente (Male ou Female)
- <b>SeniorCitizen:</b> Idoso ou não (1 ou 0)
- <b>Partner:</b> Possui um parceiro(a) (Yes ou No)
- <b>Dependents:</b> Possui dependentes(a) (Yes ou No)
- <b>tenure:</b> Número de meses que o cliente ficou com a companhia
- <b>PhoneService:</b> Possui serviço de telefone (Yes ou No)
- <b>Contract:</b> Tipo de contrato (Month-to-month, One year ou Two year)
- <b>PaperlessBilling:</b> Cobrança de conta sem papel (Yes ou No)
- <b>PaymentMethod:</b> Método de pagamento (Electronic check, Mailed check, Bank transfer ou Credit card)
- <b>MonthlyCharges:</b> Recarregamento mensal
- <b>TotalCharges:</b> Recarregamento total
- <b>Churn:</b> Cliente deixou a companhia (churn) (Yes ou No)
- <b>MultipleLines:</b> Possui múltiplas linhas de telefone (No phone service, No ou Yes)
- <b>InternetService:</b> Serviço de internet (DSL, Fiber optic, No)
- <b>OnlineSecurity:</b> Serviço de segurança online (No, Yes ou No internet service)
- <b>OnlineBackup:</b> Serviço de backup online (No, Yes ou No internet service)
- <b>DeviceProtection:</b> Proteção do dispositivo (No, Yes ou No internet service)
- <b>TechSupport:</b> Assistência técnica (No, Yes ou No internet service)
- <b>StreamingTV:</b> Possui serviço de streaming (No, Yes ou No internet service)
- <b>StreamingMovies:</b> Possui serviço de filmes em streaming (No, Yes ou No internet service)

```python
#Importando as bibliotecas

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from pandasql import sqldf
import statistics as sts
import warnings
from IPython.display import Markdown
from math import floor


warnings.filterwarnings('ignore')
sns.set_style("whitegrid")
plt.style.use('ggplot')
colors = ['royalblue','midnightblue']
```

```python
#Função para auxiliar na analise do df

def df_informations(df):
    df_info = pd.DataFrame({'Not Null': df.notnull().count(),
                'Null': df.isnull().sum(),
                'Perce Null': df.isnull().sum(),
                'Unique': df.nunique(),
                'Dtype': df.dtypes
                })

    df_dtype = pd.DataFrame(df_info['Dtype'].value_counts())
    df_dtype['Perce'] = round(df_dtype['Dtype'] / df_dtype['Dtype'].sum(), 2)

    text = f'A base tem {df.shape[0]} linhas e {df.shape[1]} Colunas. destes, temos:'

    df_info = df_info.style.background_gradient(cmap='jet', subset=['Perce Null']).format({'Perce Null': '{:.2f}'})
    df_dtype = df_dtype.style.background_gradient(cmap='YlGn', subset=['Perce']).format({'Perce': '{:.2%}'})

    display(Markdown("<H4 style='text-align:left;float:lfet;'>Informação sobre a base"))
    display(Markdown(f'<H5> {text}'))
    display(df_info)
    display(Markdown("<H4 style='text-align:left;float:lfet;'>Os tipos que nos temos:"))
    display(df_dtype)
```

```python
#Importando as bases de dados

churn_data = pd.read_csv(r'C:\Users\55359\Desktop\Projeto Chrun\churn_data.csv')
customer_data = pd.read_csv(r'C:\Users\55359\Desktop\Projeto Chrun\customer_data.csv')
internet_data = pd.read_csv(r'C:\Users\55359\Desktop\Projeto Chrun\internet_data.csv')
```

```python
#Visualizando as 5 primeiras linhas bases de dados
display(customer_data.head())
display(churn_data.head())
display(internet_data.head())
```

# Explorando a base de dados

```python
#Criando uma nova coluna com o total de planos que o cliente tem com a empresa.

cols = ['MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

internet_data['Total_de_planos'] = 0
for i, row in internet_data.iterrows():
    count = 0
    for col in cols:
        if row[col] != "No":
            count += 1
    internet_data.at[i, 'Total_de_planos'] = count
```

#### Notei que as 3 bases de dados contem informações diferentes dos mesmos clientes, então resolvi transformar as 3 bases em 1, com isso poderei explorar a base com mais eficiência e realizar o melhor tratamento.

```python
#Juntando as bases de dados
df = pd.merge(customer_data, churn_data, on='customerID', how='left')
df = pd.merge(df, internet_data, on='customerID', how='left')
#Agora df virou nossa base principal com todos os dados juntos.

#O df tem muitas colunas, então usarei os comandos abaixo para expandir o df.
pd.set_option('display.max_columns', None) # coluna removendo reticencias
pd.set_option('display.max_rows', None) # linha Removendo reticencias

df.head()
```

```python
vazios = df.loc[df['TotalCharges'].str.contains(' ')]
display(vazios['TotalCharges'].reset_index(False))
#Note que na coluna "TotalCharges" tem 11 linhas que aparentam estarem vazias, porem o demilitador espaço esta ocupando
```

```python
#irei tratar esse erro
#substituindo por vazio, mais para frente eu analiso como tratarei esse erro efetivamente.
df['TotalCharges'] = df['TotalCharges'].replace(' ', None)
```

```python
#Importando a função que criamos para auxiliar na analise da base de dados
df_informations(df)
```

#### Nessa análise inicial da base de dados noteis alguns pontos:
- A única coluna que possui valores nullos é a TotalCharges, total de 11 valores nullos
- O dtype da maioria das colunas estão corretos

#### Modificações que iremos realizar nas colunas:

- Substituir os valores da coluna SeniorCitizen de 1 para YES e de 0 para NO
- Alterar o tipo da colunas TotalCharges para tipo Float

```python
#Substitui os valores 1 por Yes e 0 por NO e automaticamente já foi convertido em texto.
df['SeniorCitizen'] = df['SeniorCitizen'].replace(1, 'Yes')
df['SeniorCitizen'] = df['SeniorCitizen'].replace(0, 'No')

#Alterando tipo da coluna TotalCharges para float
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
# Deletando a coluna customerID, não irá somar em nada na minha análise
df = df.drop(columns = 'customerID', axis = 1)
```

```python
#Informações estatisticas das colunas numericas
df.describe().T
```

```python
#Separando colunas numéricas e categoricas
colunas_numericas = df.select_dtypes(include=['float', 'int']).columns.tolist()
colunas_categoricas = df.select_dtypes(include=['object']).columns.tolist()
```

```python
#Analisando a distribuição das colunas numéricas
fig, ax = plt.subplots(4, 2, figsize=(12,10))

for i in range(1,8,2):
    plt.subplot(4,2,i)
    sns.histplot(data=df[colunas_numericas[floor(i/2)]],kde=True,palette = colors)
    title = colunas_numericas[floor(i/2)] + " distribution"
    plt.title(title)
    
    plt.subplot(4,2,i+1)
    sns.boxplot(x = colunas_numericas[floor(i/2)], data = df,palette = colors)
    title = colunas_numericas[floor(i/2)] + " box plot"
    plt.title(title)
    
plt.tight_layout()
```

### Observando a plotagem acima encontrei 2 pontos:
- O histograma do tenure é bimodal,ou seja tem dois picos: 1º pico de 0 até aproximadamente 15 meses, 2º pico de  até aproximadamente 10 meses.
- O histograma do TotalCharges é destocido para direita indica a ocorrência de altos valores com baixa frequência.
- No total de planos temos uma grande concentração em 6 planos.

```python
# Plotando gráficos com o total de observações das variáveis categóricas
fig, ax = plt.subplots(4, 4, figsize = (16, 24))

for k in range(len(colunas_categoricas)-1):
    plt.subplot(4, 4, k+1)
    feature = colunas_categoricas[k]
    sns.barplot(x = df[feature].value_counts().index, y = df[feature].value_counts().values, palette = colors)
    plt.title(feature + " proportion")
    if feature == "PaymentMethod":
        plt.xticks(rotation = 45)
```

### Observando a plotagem acima encontrei 3 pontos:
- Não temos uma desigualdade de genero em relação aos cliente da empres
- Mais da metade as pessoas preferem o tipo de contrato month-to-month
- Apenas 1869 pessoas deram Churn de um total de 7042, isso representa 26% do total.


```python
# Plotando gráficos da distribuição das variáveis relacionando com o Churn
fig, ax = plt.subplots(4, 2, figsize = (16, 24)) 

for k in range(1, 8, 2):
    plt.subplot(4, 2, k)
    sns.kdeplot(colunas_numericas[floor(k/2)], data = df, hue = "Churn", palette = colors)
    title = colunas_numericas[floor(k/2)] + " distribution"
    plt.title(title)
    
    plt.subplot(4, 2, k+1)
    sns.boxplot(x = "Churn", y = colunas_numericas[floor(k/2)], data = df, palette = colors)
    title = colunas_numericas[floor(k/2)] + " box plot"
    plt.title(title)
```

### Observando a plotagem acima encontrei 3 pontos:

- No tenure o casos que deram chrun entá concentrada a esquerda, ou seja pouco tempo com a empresa. O boxplot contem alguns outliers e é notavel a distancia entre o terceiro quartil e o maximo. Enquanto a Distribuição dos casos que não deram churn esta bem normalizada.

- No TotalCharges temos como já era esperado baseado no histograma em uma das plotagens anterior, um volume muito alto mais para esquerda ou seja valores mais baixos. Muito notavel que temos uma quantidade expressiva de outliers nos casos que deram churn.
